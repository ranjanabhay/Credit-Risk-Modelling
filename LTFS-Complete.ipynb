# Converting Branch Id,Supplier Id etc. as Predictor

```python
branch_id_df = pd.crosstab(index=train_data.branch_id,columns=train_data.loan_default).rename(columns={0:'non_event',1:'event'})
branch_id_df.columns.names=[""]
branch_id_df.reset_index(drop=False,inplace=True)
branch_id_df['per_non_event'] = branch_id_df['non_event']/branch_id_df.non_event.sum()
branch_id_df['per_event'] = branch_id_df['event']/branch_id_df.event.sum()
branch_id_df['woe'] = np.log(branch_id_df.per_non_event/branch_id_df.per_event)
branch_id_df['iv'] = (branch_id_df.per_non_event - branch_id_df.per_event)*branch_id_df.woe
print(branch_id_df.iv.sum())
branch_id_df['branch_id'] = branch_id_df['branch_id'].astype("object")
train_data = pd.merge(left=train_data,right=branch_id_df[['branch_id','woe']],on=['branch_id'],how='left')
```

```python
supplier_id_df = pd.crosstab(index=train_data.supplier_id,columns=train_data.loan_default).rename(columns={0:'non_event',1:'event'})
supplier_id_df.columns.names=[""]
supplier_id_df.reset_index(drop=False,inplace=True)
supplier_id_df['per_non_event'] = supplier_id_df['non_event']/supplier_id_df.non_event.sum()
supplier_id_df['per_event'] = supplier_id_df['event']/supplier_id_df.event.sum()
def calculate_woe(df):
    if df['per_non_event'] ==0 or df['per_event']==0:
        return np.log((df['per_non_event']+0.5)/(df['per_event']+0.5))
    elif df['per_non_event']>0 and df['per_event']>0:
        return np.log(df['per_non_event']/df['per_event'])

supplier_id_df['s_woe'] = supplier_id_df[['per_non_event','per_event']].apply(calculate_woe,axis='columns')
supplier_id_df['s_iv'] = (supplier_id_df.per_non_event - supplier_id_df.per_event)*supplier_id_df.s_woe
print(supplier_id_df.s_iv.sum())
supplier_id_df['supplier_id'] = supplier_id_df['supplier_id'].astype("object")
train_data = pd.merge(left=train_data,right=supplier_id_df[['supplier_id','s_woe']],on=['supplier_id'],how='left')
```

```python
manufacturer_id_df = pd.crosstab(index=train_data.manufacturer_id,columns=train_data.loan_default).rename(columns={0:'non_event',1:'event'})
manufacturer_id_df.columns.names=[""]
manufacturer_id_df.reset_index(drop=False,inplace=True)
manufacturer_id_df['per_non_event'] = manufacturer_id_df['non_event']/manufacturer_id_df.non_event.sum()
manufacturer_id_df['per_event'] = manufacturer_id_df['event']/manufacturer_id_df.event.sum()
def calculate_woe(df):
    if df['per_non_event'] ==0 or df['per_event']==0:
        return np.log((df['per_non_event']+0.5)/(df['per_event']+0.5))
    elif df['per_non_event']>0 and df['per_event']>0:
        return np.log(df['per_non_event']/df['per_event'])

manufacturer_id_df['m_woe'] = manufacturer_id_df[['per_non_event','per_event']].apply(calculate_woe,axis='columns')
manufacturer_id_df['m_iv'] = (manufacturer_id_df.per_non_event - manufacturer_id_df.per_event)*manufacturer_id_df.m_woe
print(manufacturer_id_df.m_iv.sum())
manufacturer_id_df['manufacturer_id'] = manufacturer_id_df['manufacturer_id'].astype("object")
train_data = pd.merge(left=train_data,right=manufacturer_id_df[['manufacturer_id','m_woe']],on=['manufacturer_id'],how='left')
```

```python
current_pincode_id_df = pd.crosstab(index=train_data.current_pincode_id,columns=train_data.loan_default).rename(columns={0:'non_event',1:'event'})
current_pincode_id_df.columns.names=[""]
current_pincode_id_df.reset_index(drop=False,inplace=True)
current_pincode_id_df['per_non_event'] = current_pincode_id_df['non_event']/current_pincode_id_df.non_event.sum()
current_pincode_id_df['per_event'] = current_pincode_id_df['event']/current_pincode_id_df.event.sum()
def calculate_woe(df):
    if df['per_non_event'] ==0 or df['per_event']==0:
        return np.log((df['per_non_event']+0.5)/(df['per_event']+0.5))
    elif df['per_non_event']>0 and df['per_event']>0:
        return np.log(df['per_non_event']/df['per_event'])

current_pincode_id_df['p_woe'] = current_pincode_id_df[['per_non_event','per_event']].apply(calculate_woe,axis='columns')
current_pincode_id_df['p_iv'] = (current_pincode_id_df.per_non_event - current_pincode_id_df.per_event)*current_pincode_id_df.p_woe
print(current_pincode_id_df.p_iv.sum())
current_pincode_id_df['current_pincode_id'] = current_pincode_id_df['current_pincode_id'].astype("object")
train_data = pd.merge(left=train_data,right=current_pincode_id_df[['current_pincode_id','p_woe']],on=['current_pincode_id'],how='left')
```

```python
state_id_df = pd.crosstab(index=train_data.state_id,columns=train_data.loan_default).rename(columns={0:'non_event',1:'event'})
state_id_df.columns.names=[""]
state_id_df.reset_index(drop=False,inplace=True)
state_id_df['per_non_event'] = state_id_df['non_event']/state_id_df.non_event.sum()
state_id_df['per_event'] = state_id_df['event']/state_id_df.event.sum()
def calculate_woe(df):
    if df['per_non_event'] ==0 or df['per_event']==0:
        return np.log((df['per_non_event']+0.5)/(df['per_event']+0.5))
    elif df['per_non_event']>0 and df['per_event']>0:
        return np.log(df['per_non_event']/df['per_event'])

state_id_df['st_woe'] = state_id_df[['per_non_event','per_event']].apply(calculate_woe,axis='columns')
state_id_df['st_iv'] = (state_id_df.per_non_event - state_id_df.per_event)*state_id_df.st_woe
print(state_id_df.st_iv.sum())
state_id_df['state_id'] = state_id_df['state_id'].astype("object")
train_data = pd.merge(left=train_data,right=state_id_df[['state_id','st_woe']],on=['state_id'],how='left')
```
